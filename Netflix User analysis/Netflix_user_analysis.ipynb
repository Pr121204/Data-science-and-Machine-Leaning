```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Netflix Userbase.csv')
```

```python
df.head()
```

```python
df = df.drop('User ID', axis = 1)
```

```python
df['Subscription Type'].unique()
```

```python
df['Device'].unique()
```

```python
df['Plan Duration'].unique()
```

```python
df = df.drop('Plan Duration', axis = 1)
```

```python
df["Join Date"] = pd.to_datetime(df["Join Date"])
df["Last Payment Date"] = pd.to_datetime(df["Last Payment Date"])
```

```python
sns.set(style="darkgrid")
sns.histplot(df['Subscription Type'])
```

```python
sns.countplot(x = df['Subscription Type'], hue = df['Gender'])
```

```python
sns.countplot(x = df['Device'], hue = df['Gender'])
```

```python
 df.groupby('Subscription Type')['Monthly Revenue'].agg('sum').plot(kind = 'bar')
```

```python
df['Subscription Type'].replace(['Basic', 'Premium', 'Standard'], [1, 2, 3], inplace = True)
```

```python
df['Device'].replace(['Smartphone', 'Tablet', 'Smart TV', 'Laptop'], [1, 2, 3, 4], inplace = True)
```

```python
df['Gender'].replace(['Male', 'Female'], [1, 2], inplace = True)
```
