```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
df = make_blobs(n_samples= 200, n_features= 2, centers= 4, cluster_std= 1.8, random_state= 101)
```

```python
df[0].shape
```

```python
plt.scatter(df[0][:,0], df[0][:,1], c = df[1], cmap = 'rainbow')
```

```python
df[1]
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 4)
```

```python
km.fit(df[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10, 6))
ax1.set_title("K means")
ax1.scatter(df[0][:,0], df[0][:,1], c = km.labels_, cmap = 'rainbow')
ax2.set_title("Actual")
ax2.scatter(df[0][:,0], df[0][:,1], c = df[1], cmap = 'rainbow')
```
